```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import matplotlib.pyplot as plt        #Visualisation
import seaborn as sns                  #Visualisation
%matplotlib inline
```

```python

titanic = pd.read_csv("https://sites.google.com/site/yasinunlu/home/research/new1/Titanic_train.csv")
titanic.head(20)
```

```python
def data_reprocessing(titanic):
  titanic = titanic.drop(['PassengerId', 'Ticket'], axis = 1)
  missing_entries = titanic[titanic.isnull().any(axis=1)]
  titanic = titanic.dropna()
  titanic['Fare']= titanic['Fare'].astype(int)
  titanic['Age']= titanic['Fare'].astype(int)
  titanic = titanic.rename(columns = {'SibSp': 'Siblings/Spouses Aboard', 'Parch': 'Parents/Children Aboard'}, inplace = False)
  titanic['Name'] = titanic['Name'].replace(['Mill','Ms','Mme'],['Miss','Miss','Mrs'])
  return titanic
titanic = data_reprocessing(titanic)
titanic.head()
```
