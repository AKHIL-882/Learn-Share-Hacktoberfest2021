```python
// defining the class name Book
public class Book{
    // int data feild naed pages that stores the # of pages
    private int pages;
    // A string data field named title that represents the tile of the book
    private String title;
    // get method to return pages
    public int getPages() {
        return pages;
    }
    // set method to set the pages
    public void setPages(int pages) {
        this.pages = pages;
    }
    // get method to get the title
    public String getTitle() {
        return title;
    }
    // set method to set the title
    public void setTitle(String title) {
        this.title = title;
    }
    // initializing the calss with pages and title
    public Book(int pages, String title) {
        super();
        this.pages = pages;
        this.title = title;

    }
    // toString method that returns book information
    @Override
    public String toString() {
        // returing the iinformation
        return "Book [pages=" + pages + ", title=" + title + "]";
    }
    // equals method that returns true and false
    public boolean equals(Book b){
    // checking the condition
    if(pages == b.pages && title.equals(b.title))
        return true;
    else
        return false;
    }
    // comapre method that compares two books
    public int compareTo(Book b){
        if(pages < b.pages)
        // retruns -1 is the first book has less pages than the second one
            return -1;
        // return 0 both the books aare same #of pages
        else if(pages == b.pages)
            return 0;
        else
        // return 1 is the first book has more pages than the second one.
            return 1;
        }

}

```
