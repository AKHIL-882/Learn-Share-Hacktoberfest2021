```python
// importing the libraries
import java.util.Scanner;// for getting input from user
// defining the main class
public class Main{
    // initializing the main method
    public static void main(String[] args){
         // defining the scanner class
         Scanner sc = new Scanner(System.in);
         System.out.print("Enter a lowercase or an uppercase letter: ");
         // variable to store user character
         char userCharacter ;
         // looping the to check the character
         while (true){
              userCharacter = sc.next().charAt(0);
              if (Character.isLetter(userCharacter)){
                  break;
              }
              // is the inputed Character not string
              else{
                  System.out.println("Invalid Input");
                  break;
              }
         }
         // passing the argument Character and converting it into lowercase
         switch( Character.toLowerCase(userCharacter)){
               // Displaying 2 for A-B-C 
               case 'a':
               case 'b':
               case 'c':
               System.out.println("The corresponding number is 2 ");
               break;
             
               // Displaying 3 for D-E-F
               case 'd':
               case 'e':
               case 'f':
               System.out.println("The corresponding number is 3 ");
               break;
             
               // Displaying 4 for G-H-I
               case 'g':
               case 'h':
               case 'i':
               System.out.println("The corresponding number is 4 ");
               break;
             
               // Displaying 5 for J-K-L
               case 'j':
               case 'k':
               case 'l':
               System.out.println("The corresponding number is 5 ");
               break;
             
               // Displaying 6 for M-N-O
               case 'm':
               case 'n':
               case 'o':
               System.out.println("The corresponding number is 6 ");
               break;
             
               // Displaying 7 for P-Q-R-S
               case 'p':
               case 'q':
               case 'r':
               case 's':
               System.out.println("The corresponding number is 7 ");
               break;
             
             
               // Displaying 8 for T-U-V
               case 't':
               case 'u':
               case 'v':
               System.out.println("The corresponding number is 8 ");
               break;
             
               // Displaying 9 for W-X-Y-Z
               case 'w':
               case 'x':
               case 'y':
               case 'z':
               System.out.println("The corresponding number is 9 ");
               break;
          }
      }
}

```
