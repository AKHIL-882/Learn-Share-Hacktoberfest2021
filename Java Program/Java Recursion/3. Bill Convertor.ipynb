```python
package billConveror;
//importing the libraries
import java.io.*;
import java.util.*;
//initializing the main function
public class billConvertor{
 public static void main(String args[]){
     // initializing the variables
     int userPeakMinutes, userEveningMinutes;
     // bill per call
     double bill = 19.99;
     // checking whether the arguments are greater or less than 2
     if(args.length < 2 || args.length>2){
         // displaying the error message to the user
         System.out.println("\nPlease provide peak minutes and evening minutes - which are of two values: ");
         System.exit(1);
     }
     // defining the try and catch 
     try {
         // converting the string type arguments into integers
         userPeakMinutes = Integer.parseInt(args[0]);
         userEveningMinutes = Integer.parseInt(args[1]);
         // checking whether the peak minutes greater than 200 or not
         if (userPeakMinutes > 200)
             // considering the peak minutes to calculate the bill
             userPeakMinutes = userPeakMinutes - 200;
         else userPeakMinutes = 0;
         // checking whether the peak minutes greater than 500 or not
         if (userEveningMinutes > 500) 
             // considering the peak minutes to calculate the bill
             userEveningMinutes = userEveningMinutes - 500;
         else userEveningMinutes = 0;
         
         // considering the extra call timings
         bill = bill + userPeakMinutes * 0.30 + userEveningMinutes * 0.15;
         // displaying the Monthly bill
         System.out.println("\nMonthly bill = $" + bill);

     }
     // initializing the catch block
     catch (NumberFormatException numberFormatException) {
         // if arguments are not integers print error message and exit
         System.out.println("\nBoth arguments must be integers.");
         System.exit(1);
     }         
   

 }
}
```
