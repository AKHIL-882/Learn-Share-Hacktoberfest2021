```python
// importing the libraries
import java.util.Scanner;
// initializing the main class
public class Main {
    // declaring the main function
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        // getting the cents from the user
        int userCents = scanner.nextInt();
        // checking the cents are 0 
        if (userCents <= 0) {
            System.out.println("No change");
        } else {
            // converting into dollars
            int dollars = userCents / 100;
            userCents %= 100;
             // converting into quarters
            int quarters = userCents / 25;
            userCents %= 25;
            // converting into dimes
            int dimes = userCents / 10;
            userCents %= 10;
            // converting into nickels
            int nickels = userCents / 5;
            // converting into pennies
            int pennies = userCents % 5;
            // displaying the result
            if (dollars != 0)
                System.out.println(dollars + (dollars == 1 ? " Dollar" : " Dollars"));
            if (quarters != 0) 
                System.out.println(quarters + (quarters == 1 ? " Quarter" : " Quarters"));
            if (dimes != 0) 
                System.out.println(dimes + (dimes == 1 ? " Dime" : " Dimes"));
            if (nickels != 0) 
                System.out.println(nickels + (nickels == 1 ? " Nickel" : " Nickels"));
            if (pennies != 0) 
                System.out.println(pennies + (pennies == 1 ? " Penny" : " Pennies"));
        }
    }
}
```
