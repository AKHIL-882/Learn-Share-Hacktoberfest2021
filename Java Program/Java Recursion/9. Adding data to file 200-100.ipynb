```python
// importing the libraries
import java.io.FileWriter;
import java.io.IOException;
// defining the public class
public class Main {
        // setting the file path
        // Note: Remane the file name if you wish.
        static final String filePath = "MyNumbers-YOUR_NAME-YOUR_ID.txt";
        // declaring the main function
        public static void main(String[] args) {
                // initializing the try catch block
                try {
                        // using the FileWriter class
                        FileWriter myWriter = new FileWriter(filePath);
                        String data = "";
                        // loop to add data to the file
                        for (int i = 200; i >= 100; i -= 2)
                                data += i + ", ";
                        // adding the data to the file
                        myWriter.write(data);
                        myWriter.close();
                        // display the msg to the user
                        System.out.println("Data Added Successfully");
                }
                // displaying error if encounterred
                catch (IOException e) {
                        System.out.println("An error occurred : " + e.getMessage());
                }
        }
}
```
