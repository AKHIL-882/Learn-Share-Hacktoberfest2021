```python
#include <stdio.h>
//global variables
int count=0;

int main()
{
    // declare the constant value of 4 for the array size
    const int NUM_VALS = 4;
    // declaring the array with the constant size
    int userValues[NUM_VALS],i;
    //declaring the variables
    int matchValue;
    
    // user input of match Value
    printf("Enter the matchValue: \t");
    scanf("%d",&matchValue);
    
    printf("Enter the userValues Elements:\n");
    // for loop to get the array of elements for the user
    for(i=0;i<NUM_VALS;i++){
        scanf("%d",&userValues[i]);
    }
    
 
    // condition for checking the match cases
    for(i=0;i<NUM_VALS;i++){
        if(userValues[i]==matchValue)
            count++;  // count will be incremented when there is a match
    }
    
    
    // no matches
    if(count==0){
        printf("\nNo match Found");
        printf("The Number of match values are: \t%d", count);
    }
    else
    // printing the number of match cases
       printf("The Number of match values are: \t%d", count);
}
```
