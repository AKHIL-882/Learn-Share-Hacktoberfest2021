```python
// importing the libraries
#include <iostream>
#include <fstream>
#include <string>
#include <stdio.h>
#include <bits/stdc++.h>
using namespace std;
// initializing the main function
int main()
{
    // displaying information
    cout << "Welcome to Wake's ID Processing Agency" << endl;
    cout << "What is the name of your file? ";
    // declaing the file name
    string userFileName;
    cin >> userFileName;
    // reading and opening the file
    ifstream file;
    file.open(userFileName);
    // initializing the smallest int value
    int defineMaximumId = INT_MIN;
    // initializing the largest int value
    int defineMinimumId = INT_MAX;
    // variable to store the sum, sum of even, count of id's count of even id's
    int sumId = 0,sumEvenID = 0,userNumberId = 0,numEvenID = 0;
    // looping the file to get the context
    while (!file.eof())
    {
        int ID;
        // getting the first and last names from the file
        string userFirstName;
        string userLastName;
        file >> userFirstName;
        file >> userLastName;
        file >> ID;
        // increment id count when you see a new id
        userNumberId++;
        // sum of id's
        sumId += ID;
        // if id is greater than largest id then change the id
        if (ID > defineMaximumId)
            defineMaximumId = ID;
        // if id is less than smallest id then change the id
        if (ID < defineMinimumId)
            defineMinimumId = ID;
        // checking if the id is even or not
        if (ID % 2 == 0){
            numEvenID++;
            // sum of even id's
            sumEvenID += ID;
        }
    }
    // displaying number of students in the file
    cout << "The number of Students in the file is " << userNumberId << endl;
    // displaying even and odd id's count
    cout << "There are " << userNumberId - numEvenID << " odd IDs and " << numEvenID << " even IDs" << endl;
    // displaying sum of odd id's
    cout << "The sum of the odd IDs is " << sumId - sumEvenID << endl;
    // displaying sum of even id's
    cout << "The sum of the even IDs is " << sumEvenID << endl;
    // displaying sum of all id's in the file
    cout << "The sum of all the IDs is " << sumId << endl;
    // displaying the largest id
    cout << "The largest ID is " << defineMaximumId << endl;
    // displaying the smallest id
    cout << "The smallest ID is " << defineMinimumId << endl;
    // displaying the average of largest and smallest
    cout << "The average of " << defineMaximumId << " and " << defineMinimumId << " is ";
    // print the average of largest and smallest id
    printf("%.2f\n", (defineMaximumId + defineMinimumId) / 2.0);
    // closing the file
    file.close();
}
```
