**C++ Program to find the factorial of a number with recursion**

```python
// importing the libraries
#include <iostream>
using namespace std;
  
// function to find factorial
unsigned int factorial(unsigned int n)
{
    if (n == 0)
        return 1;
    return n * factorial(n - 1);
}
  
// declare main function
int main()
{
    // initialize the variables
    int num;
    cout << "Enter a positive integer: ";
    cin >> num;
    // passing num into factorial function as an argument
    cout << "Factorial of "
         << num << " is " << factorial(num) << endl;
    return 0;
}
```

**C++ program to find the factorial of a number without recursion**

```python
// importing the libraries
#include <iostream>
using namespace std;

// declare main function
int main() {
    
    // initializing the varibles
    int n;
    long double number = 1.0;

    // User input to find the factorial of a number
    cout << "Enter a positive integer: ";
    cin >> n;

    // Condition to check negative numbers
    if (n < 0)
        cout << "Please Enter positive numbers";
        
        // Condition to check positive numbers
    else {
        for(int i = 1; i <= n; ++i) {
            number *= i;   // here will we multiply the value with the previous number by incrementing the i value. 
        
        }
        cout << "Factorial of " << n << " = " << number;    
    }

    return 0;
}
```
