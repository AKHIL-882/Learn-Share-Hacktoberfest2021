```python
// importing the libraries
#include <bits/stdc++.h>
using namespace std;
// initializing the main function
int main(){
        // defining the vector to store student student Id Numbers
        vector<string> studentIdNumbers;
        // declaring object of fstream class
        fstream file;
        string fileName = "studentDetails.txt";
        // opening the file
        file.open(fileName.c_str(), ios::in);   
        // if file is not found
        if(!file.is_open()){
                cout<<"FileError!! File Not Found";
                exit(1);
        }
        // while loop to iterate over each line 
        // in the file
        while(!file.eof()){
                string idNumber;
                // getting the number from the file
                file >> idNumber;
                // reversing the number
                reverse(idNumber.begin(), idNumber.end());
                // adding the number to vector
                studentIdNumbers.push_back(idNumber);
        }
        // sorting the vector
        sort(studentIdNumbers.begin(), studentIdNumbers.end());
        int finalResult = 0;
        // for loop to traverse over the vector
        for(int i=0; i < studentIdNumbers.size() - 1; i++){
                // declaring the local variable 
                // to store the count value
                int count = 0;
                for(int j=0; j<studentIdNumbers[i].size(); j++){
                        // condition to check if there are any common number
                        if(studentIdNumbers[i][j] == studentIdNumbers[i+1][j])
                                // incrementing count value
                                count++;
                        else
                                break;
                }
                // storing common values
                finalResult = max(count, finalResult);
        }
        // displaying the result
        cout<<"Number of digits which is used to uniquely identify student number are: "<<(finalResult + 1);
        return 0;                       
}
```
