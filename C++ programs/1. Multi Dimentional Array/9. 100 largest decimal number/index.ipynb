```python
// importing the libraries
#include <iostream>
#include <list>
#include <algorithm>
#include <fstream>
#include <string>
using namespace std;

// Function to add all numbers
void addAllElementsInFile(list<int> &numsInFile){
    // reading a file
    fstream userDefinedFile; 
    // opening the  file 
    userDefinedFile.open("nums.txt",ios::in);
    // for loop for push back data
    for (int i = 0; i < 50; i++){
        numsInFile.push_back(0);
    }

    // setting carry
    int carryIFOverflow = 0;
    // getting the start number
    string startNumberInFile; 
    // checking the conidtion
    while (userDefinedFile >> startNumberInFile){
        // traversing the list
        auto reverseListNum = numsInFile.begin();

        // adding the number to the list in reverse order
        for (int i = startNumberInFile.length() - 1; (i >= 0) && (reverseListNum != numsInFile.end()); reverseListNum++, i--){
            // converting to integer
            int digitsUserInFile = (startNumberInFile[i] - '0'); 
            // sum of the digits
            int sumValues = carryIFOverflow + digitsUserInFile + (*reverseListNum);
            // getting the carry
            carryIFOverflow = sumValues / 10;
            // getting the new digit
            int newdigitsUserInFile = sumValues % 10;
            // replacing the values
            (*reverseListNum) = newdigitsUserInFile;
        }

        // checking the carry is generated 
        while (carryIFOverflow != 0){
            // add carry value to the end of the list
            if (reverseListNum == numsInFile.end()){
                // pushing the elements back
                numsInFile.push_back(1); 
                carryIFOverflow = 0;        
            }
            else{
                // sum of all values
                int sumValues = *reverseListNum + carryIFOverflow; 
                // replaing the sum values
                *reverseListNum = sumValues % 10;  
                //replacing the current value to the sum values
                carryIFOverflow = sumValues / 10;   
                // getting the carry and traverse to next element
                reverseListNum++;                  
            }
        }
    }
}


// funtion to display the elements
void displayElements(list<int> numsInFile){
    cout<<"The sum of all elements are: ";
    // printing the values in reverse order
    for (auto i = numsInFile.rbegin(); i != numsInFile.rend(); i++){
        cout << *i;
    }
}

int main()
{
    // declaring the list
    list<int> numsInFile;
    // calling the function to add the values in file
    addAllElementsInFile(numsInFile);
    // displaying the result
    displayElements(numsInFile);
    return 0;
}
```
