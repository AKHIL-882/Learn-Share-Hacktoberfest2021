```python

# defining the function
def countOdds(userList):
  # variable to store the count value
  count = 0
  # iterating over the elements
  for i in range(len(userList)):
    # condition for getting the odd numbers
    if(userList[i]%2!=0):
      # if odd number found
      # incrementing the cound
      count+=1
  # returing the count value
  return count

# getting the result
print(countOdds([1,2,3]))
print(countOdds([]))
print(countOdds([3,3,5]))
```
