```python
# displaying the service name
print('Chanuceys Vet Services')
# creating a set of services 
veterinaryService = {'ExamEdit': 45,'Vaccinations': 32,'Trim Nails': 8,'Bath': 15 }
print('\n')

# displaying the service and cost to the user in menu driven
for i,j in veterinaryService.items():
  # displaying the menu
  print('{}: {}'.format(i,j))

print('\n')
  
# getting the input of two services from user
userServiceOne = input('Select first service: ')
userServiceTwo = input('Select second service: ')

print('\n')
print("Chauncey's Vet Invoice")
# variable to store the sum of services
totalBill = 0
# checking if the user entered is none or not
if(userServiceOne == 'None' or userServiceOne =='none'):
  print('Service 1 - None: 0')
  # if condition is true, we have to add 0 to the total bill
  totalBill+=0
else:
  # mapping the user entered value with the values in set
  print("Service 1 - {}:{}".format(userServiceOne,veterinaryService[userServiceOne]))
   # displaying the total service and bill
  totalBill+=veterinaryService[userServiceOne]
  
# checking if the user entered is none or not
if userServiceTwo == 'None' or userServiceTwo =='none':
  print('Service 2 - None: 0')
  # if condition is true, we have to add 0 to the total bill
  totalBill+=0
else:
  # mapping the user entered value with the values in set
  print("Service 2 - {}:{}".format(userServiceTwo,veterinaryService[userServiceTwo]))
   # displaying the total service and bill
  totalBill+=veterinaryService[userServiceTwo]
# displaying the total bill to the user
print("Total :",totalBill)
```

```python
class BankAccount:
    def __init__(self,accountNumber, name, balance):
        self.accountNumber = accountNumber
        self.name = name
        self.balance = balance
    def Deposit(self , d ):
        self.balance = self.balance + d
    def Withdrawal(self , w):
        if(self.balance < w):
            print("impossible operation! Insufficient balance !")
        else:
            self.balance = self.balance - w
    def bankFees(self):
        self.balance = (95/100)*self.balance
    def display(self):
        print("Account Number : " , self.accountNumber)
        print("Account Name : " , self.name)
        print("Account Balance : " , self.balance , " $")
        
bankAccount = int(input("Enter you bank account number: "))
userName = input("Enter your name: ")
userAmount = int(input("Enter your amount: "))
newAccount = BankAccount(bankAccount,userName,userAmount)
userWithdrawAmount = int(input("Enter the amount to be withdraw: "))
newAccount.Withdrawal(userWithdrawAmount)
userDepositeAmount = int(input("Enter the amount to be deposite: "))
newAccount.Deposit(userDepositeAmount)
newAccount.display()
```

```python
# defining the function
def make_index(words_on_page):
  # declaring empty set
  new_dictonary = {}
  # iterating over the element
  for i in words_on_page:
    # tracing the value of keys
    for j in words_on_page.get(i):
      # if the key is not in new dictonary 
      if j not in new_dictonary:
        # add the key
        new_dictonary[j] = [i]
      else:
        # append a value
        new_dictonary[j].append(i)
  # return a dictionary to the user
  return new_dictonary

# input from user
input_dict={1: ['hi', 'there', 'fred'], 2: ['there', 'we', 'go'], 3: ['fred', 'was', 'there']}
# displaying the result to the user
print(make_index(input_dict))
```
