### **Python Program to convert Kilometers into miles, feet and inches**

qu1.png

```python
import math

km = float(input("Enter the value in kilometers: "))

mi = km * 0.621371

separated_mi = math.modf(mi)
foot = separated_mi[0] * 5280


separated_foot = math.modf(foot)
inches = separated_foot[0] * 12

format_inches = "{:.2f}".format(inches)

print ("{} kilometers equals {} miles , {} feet, and  {} inches"
       .format(int(km), int(mi) ,int(foot), format_inches))

```
