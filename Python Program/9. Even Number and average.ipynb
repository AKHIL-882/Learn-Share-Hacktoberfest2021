```python
# declare an empty list
integers=[]
# variable to sum the even numbers
evenSum=0   
# number of elements     
number = 10
# num variable to calculate the no.of even numbers
num = 0
# for loop for taking the integers
for i in range(1, number+1):
    value=int(input("Please enter the value of %d Elements: " %i))
    # appending value to the list integers
    integers.append(value)
# for loop to find the sum of even numbers
for j in range(number):
    # finding sum of numbers if the value if greater then 0
    if(integers[j]>=0 and integers[j]%2==0):
       # calculating the sum of even numbers
        evenSum=evenSum+integers[j]
        # calculating the no.of even numbers
        num= num+1
# printing the even and average of all numbers
print("\n The sum of even numbers in the list: ",evenSum)
print("\n The average of even numbers: ", evenSum/num)
print ("\n The average of with 2 decimal places: ",end="")
print ("{0:.2f}".format(evenSum/num))
```
