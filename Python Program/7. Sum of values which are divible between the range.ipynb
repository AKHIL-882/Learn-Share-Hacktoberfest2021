```python
# defining the function result with 3 arguments minimum, maximum, divisor values

def result(minimum, maximum, divisor):

    # initializing sum varible to add the sum of values

    sum=0

    # for loop for iterating over the numbers from minimum to maximum

    for num in range(minimum, maximum):

            # checking that the number is divisible by divisor 

            if num % divisor == 0:

              # if the number is divisible then we have to add that number to sum

              sum = sum+num

            # if the number is not divisible just pass the loop

            else:

                pass

    # return the total sum value

    return sum

# user input for minimum, maximum, divisor values

minimum, maximum, divisor = [int(x) for x in input("Enter minimum, maximum, divisor value: ").split()]

# calling the function by passing minimum, maximum, divisor values

val = result(minimum, maximum, divisor)

# printing the result

print("\n",val,"\n")
```
